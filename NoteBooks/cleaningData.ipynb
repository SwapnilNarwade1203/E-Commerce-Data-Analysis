Data Cleaning
A real analyst typically cleans data in this order:
1.Remove duplicates
2.Handle missing values
3.Fix data types
4.Standardize text/categorical values
5.Detect outliers
6.Feature engineering
7.Validate cleaned data

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../Data/train.csv")
```

```python
#Remove Duplicate
duplicates = df.duplicated(keep=False).sum()
print("\nDuplicate Rows Before:", duplicates)

df = df.drop_duplicates()

print("Duplicate Rows After:", df.duplicated().sum())
print("Shape After Removing Duplicates:", df.shape)
```

```python
#Handle Missing Values
print("\n Missing Values before Handling ",df.isnull().sum())

#numeric missing values using median
numeric_col=df.select_dtypes(include=np.number).columns
for col in numeric_col:
    if df[col].isnull().sum() > 0:
        df[col]=df[col].fillna(df[col].median())

#categorial missing values using mode
categorial_col=df.select_dtypes(include='object').columns
for col in categorial_col:
    if df[col].isnull().sum() >0:
        df[col]=df[col].fillna(df[col].mode())

print("\n Missing Values before Handling ",df.isnull().sum())
```

```python
# 3. Fix Data Types
print("\n===== DATA TYPES BEFORE =====")
print(df.dtypes)
# Convert Dates
date_cols = ['Order Date', 'Ship Date']

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col],format='%d/%m/%Y')


# Convert IDs to Category
id_cols = ['Order ID', 'Product ID', 'Customer ID']

for col in id_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')


# Postal Code should be String
if 'Postal Code' in df.columns:
    df['Postal Code'] = df['Postal Code'].astype(str)


print("\n===== DATA TYPES AFTER =====")
print(df.dtypes)
```

```python

# ==========================================
# 5. STANDARDIZE TEXT COLUMNS
# ==========================================
text_cols = [
    'Customer Name', 'Segment', 'Country', 'City',
    'State', 'Region', 'Category', 'Sub-Category',
    'Product Name', 'Ship Mode'
]

for col in text_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.title()

print("\nText Standardization Completed")


# ==========================================
# 6. FEATURE ENGINEERING
# ==========================================
if 'Order Date' in df.columns:
    df['Order Year'] = df['Order Date'].dt.year
    df['Order Month'] = df['Order Date'].dt.month
    df['Order Day'] = df['Order Date'].dt.day
    df['Order Month Name'] = df['Order Date'].dt.month_name()

    print("\nDate Features Created")


# Shipping Duration Feature
if 'Order Date' in df.columns and 'Ship Date' in df.columns:
    df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days
    print("Shipping Days Feature Created")


# ==========================================
# 7. FINAL VALIDATION
# ==========================================
print("\n===== CLEANED DATA INFO =====")
print(df.info())

print("\nSample Cleaned Data:")
print(df.head())


# ==========================================
# 8. SAVE CLEANED DATA
# ==========================================
df.to_csv("../Data/cleaned_superstore.csv", index=False)

print("\nCleaned dataset saved successfully!")
```
