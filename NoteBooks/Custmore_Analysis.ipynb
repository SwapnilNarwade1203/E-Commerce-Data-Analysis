```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style("whitegrid")
df=pd.read_csv("../Data/cleaned_superstore.csv")
```

```python
# ==========================================
# 1. TOP 10 CUSTOMERS BY SALES
# ==========================================
top_customers = (
    df.groupby('Customer Name')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print("Top 10 Customers by Sales:")
print(top_customers)

plt.figure(figsize=(10,6))
sns.barplot(
    x=top_customers.values,
    y=top_customers.index
)
plt.title("Top 10 Customers by Sales")
plt.xlabel("Sales")
plt.ylabel("Customer")
plt.show()
```

```python
# ==========================================
# 2. MOST FREQUENT CUSTOMERS
# ==========================================
frequent_customers = (
    df.groupby('Customer Name')['Order ID']
    .count()
    .sort_values(ascending=False)
    .head(10)
)

print("\nMost Frequent Customers:")
print(frequent_customers)

plt.figure(figsize=(10,6))
sns.barplot(
    x=frequent_customers.values,
    y=frequent_customers.index
)
plt.title("Most Frequent Customers")
plt.xlabel("Number of Orders")
plt.ylabel("Customer")
plt.show()
```

```python
# ==========================================
# 3. AVERAGE REVENUE PER CUSTOMER
# ==========================================
avg_revenue_customer = (
    df.groupby('Customer Name')['Sales']
    .sum()
    .mean()
)

print("\nAverage Revenue Per Customer:", round(avg_revenue_customer, 2))
```

```python
# ==========================================
# 4. CUSTOMER REVENUE DISTRIBUTION
# ==========================================
customer_sales = df.groupby('Customer Name')['Sales'].sum()

plt.figure(figsize=(10,5))
sns.histplot(customer_sales,bins=30,kde=True)

plt.title("Customer Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```
