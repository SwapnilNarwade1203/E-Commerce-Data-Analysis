```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style("whitegrid")
df=pd.read_csv("../Data/cleaned_superstore.csv")
```

```python
# ==========================================
# 1. TOTAL SALES
# ==========================================
total_sales = df['Sales'].sum()
avg_sales = df['Sales'].mean()

print("Total Sales:", round(total_sales, 2))
print("Average Sales Per Order:", round(avg_sales, 2))
```

```python
# ==========================================
# 2. SALES BY CATEGORY
# ==========================================
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

print("\nSales By Category:")
print(category_sales)

plt.figure(figsize=(7,5))
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python

# ==========================================
# 3. SALES BY REGION
# ==========================================
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

print("\nSales By Region:")
print(region_sales)

plt.figure(figsize=(8,5))
sns.barplot(x=region_sales.index, y=region_sales.values)
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
# ==========================================
# 4. MONTHLY SALES TREND
# ==========================================
monthly_sales = df.groupby('Order Month Name')['Sales'].sum()

month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

monthly_sales = monthly_sales.reindex(month_order)

print("\nMonthly Sales:")
print(monthly_sales)

plt.figure(figsize=(12,5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# ==========================================
# 5. TOP 10 PRODUCTS
# ==========================================
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

print("\nTop 10 Products:")
print(top_products)

plt.figure(figsize=(10,6))
sns.barplot(y=top_products.index, x=top_products.values)
plt.title("Top 10 Products by Sales")
plt.xlabel("Sales")
plt.ylabel("Product")
plt.show()
```

```python
shipmode_wise_sale=df.groupby('Ship Mode')['Sales'].sum()
print("\n Shipmode Wise Sale")
print(shipmode_wise_sale)

plt.Figure(figsize=(25,5))
plt.pie(
    shipmode_wise_sale,
    labels=shipmode_wise_sale.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Shipmode Wise Sale")
plt.show()
```
