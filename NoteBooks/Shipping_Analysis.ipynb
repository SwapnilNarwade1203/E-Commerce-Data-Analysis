```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style("whitegrid")
df=pd.read_csv("../Data/cleaned_superstore.csv")
```

```python
#Average Shipping Days
avg_shipping = df['Shipping Days'].mean()

print("Average Shipping Time:", round(avg_shipping, 2), "days")
```

```python
#Shipping Time Region
shipping_region = (
    df.groupby('Region')['Shipping Days']
    .mean()
    .sort_values()
)

print(shipping_region)

plt.figure(figsize=(8,5))
sns.barplot(
    x=shipping_region.index,
    y=shipping_region.values
)

plt.title("Average Shipping Days by Region")
plt.xlabel("Region")
plt.ylabel("Shipping Days")
plt.show()
```

```python
#Shipping time by Shipping Mode
shipping_mode = (
    df.groupby('Ship Mode')['Shipping Days']
    .mean()
    .sort_values()
)

print(shipping_mode)

plt.figure(figsize=(8,5))
sns.barplot(
    x=shipping_mode.index,
    y=shipping_mode.values
)

plt.title("Shipping Performance by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Average Shipping Days")
plt.xticks(rotation=20)
plt.show()
```

```python
#Delay Order Detecation
delayed_orders = df[df['Shipping Days'] > 5]

print("Delayed Orders Count:", len(delayed_orders))
```
